```python
import matplotlib.pyplot as plt
import numpy as np

import poppy
from poppy.tests.test_matrixDFT import makedisk
import poppy.matrixDFT as matrixDFT

import math
```

```python
px = 100

pupil = makedisk(s=(px, px), c=(49.5, 49.5), r=25.0) 
#phase = np.zeros((100, 100)) 

nx, ny = (px, px)
x = np.linspace(0, 1, nx)
y = np.linspace(0, 1, ny)
y_tilt, x_tilt = np.meshgrid(x, y)


spatialwavelen = 0.1           #determines spatial frequency
center = (0, 100)              #doesn't shift the sinusoid?
offset = 0              #phase shifts, put multiples of np.pi
khat = (1, 1)                  #determines orientation

argu  = 2.0 * np.pi * (khat[0] * (x_tilt - center[0])  +  khat[1] * (y_tilt - center[1])) / spatialwavelen 
coswave = np.cos(argu - offset)

print("max=",np.max(coswave), "min=",np.min(coswave), "centre pixel value=",coswave[50,50])

phase = coswave

pupilarray = pupil * np.exp(1j * phase)
focalplane_size=100  # in l/D units
focalplane_npix=100

mft = matrixDFT.MatrixFourierTransform()
imagefield = mft.perform(pupilarray, focalplane_size, focalplane_npix)
```

```python
plt.figure(figsize=(15,7.5))
plt.subplot(2, 3, 1)
plt.title("Re(Pupil)")
plt.imshow(np.real(pupilarray))
plt.colorbar()

plt.subplot(2, 3, 2)
plt.title("Im(Pupil)")
plt.imshow(np.imag(pupilarray))
plt.colorbar()

plt.subplot(2, 3, 3)
plt.title("Pupil_phase")
plt.imshow(phase)
plt.colorbar()

plt.subplot(2, 3, 4)
plt.title("Re(Imagefield)")
plt.imshow(np.real(imagefield))
plt.colorbar()

plt.subplot(2, 3, 5)
plt.title("Im(Imagefield)")
plt.imshow(np.imag(imagefield))
plt.colorbar()

plt.subplot(2, 3, 6)
plt.title("Intensity")
plt.imshow(np.abs(imagefield)**2)
plt.colorbar()
```

Compare phases, symmetric amplitude should result in real part only, imag =0
anti-symmetric amplitude should result in real = 0, imag part only

```python
spatialwavelen = 0.1          #determines spatial frequency
center = (0, 100)            #doesn't shift the sinusoid?
offset = 0                   #phase shifts, put multiples of np.pi
khat = (1, 0)                #determines orientation

argu  = 2.0 * np.pi * (khat[0] * (x_tilt - center[0])  +  khat[1] * (y_tilt - center[1])) / spatialwavelen 

coswave = np.cos(argu - offset)
sinwave = np.sin(argu - offset)

pupilarray1 = pupil * np.exp(1j * coswave)
pupilarray2 = pupil * np.exp(1j * sinwave)

# pupilarray1 = pupil * coswave
# pupilarray2 = pupil * sinwave 

imagefield_cos = mft.perform(pupilarray1, focalplane_size, focalplane_npix)
imagefield_sin = mft.perform(pupilarray2, focalplane_size, focalplane_npix)
```

```python
plt.figure(figsize=(10,25))

plt.subplot(5, 2, 1)
plt.title("cos_phase")
plt.imshow(coswave*pupil)
plt.colorbar()

plt.subplot(5, 2, 2)
plt.title("sin_phase")
plt.imshow(sinwave*pupil)
plt.colorbar()

plt.subplot(5, 2, 3)
plt.title("1D cos phase")
plt.plot((sinwave*pupil)[:, 49], label='sinwave')
plt.plot((coswave*pupil)[:, 49], label='coswave')
plt.legend()

plt.subplot(5, 2, 4)
plt.title("1D cos phase")
plt.plot((sinwave*pupil)[49, :], label='sinwave')
plt.plot((coswave*pupil)[49, :], label='coswave')

plt.subplot(5, 2, 5)
plt.title("1D Intensity") #np.abs retunes sqrt 
plt.plot((np.abs(imagefield_cos)**2)[:, 49], label='coswave')
plt.plot((np.abs(imagefield_sin)**2)[:, 49], label ='sinwave')
plt.legend() 

plt.subplot(5, 2, 6)
plt.title("Re(Imagefield_cos)")
plt.imshow(np.real(imagefield_cos))
plt.colorbar()

plt.subplot(5, 2, 7)
plt.title("Im(Imagefield_cos)")
plt.imshow(np.imag(imagefield_cos))
plt.colorbar()

plt.subplot(5, 2, 8)
plt.title("Re(Imagefield_sin)")
plt.imshow(np.real(imagefield_sin))
plt.colorbar()

plt.subplot(5, 2, 9)
plt.title("Im(Imagefield_sin)")
plt.imshow(np.imag(imagefield_sin))
plt.colorbar()

plt.subplot(5, 2, 10)
plt.title("diff intensities")
plt.imshow((np.abs(imagefield_cos))**2 -  (np.abs(imagefield_sin))**2)
plt.colorbar()
```
